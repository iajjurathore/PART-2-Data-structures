Task 2 — Cart Operations
==

```python

cart = []


# == Write logic to add an item to the cart -----
def add_item(item_name, quantity):

    # ---Check if the item exists in menu---
    if item_name not in menu:
        print(f"'{item_name}' does not exist in the menu.")
        return

    # ---Check if item is available
    if menu[item_name]["available"] == False:
        print(f"'{item_name}' is currently unavailable.")
        return

    # ---Check if item is already in the cart
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = entry["quantity"] + quantity
            print(f"Updated '{item_name}' quantity to {entry['quantity']}.")
            return

    # ---Item not in cart yet, add new entry----
    new_entry = {
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    }
    cart.append(new_entry)
    print(f"Added '{item_name}' x{quantity} to cart.")


# --Function to REMOVE an item from the cart------
def remove_item(item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print(f"Removed '{item_name}' from cart.")
            return

    print(f"'{item_name}' is not in the cart.")


# -- Function to UPDATE quantity of an item -----
def update_quantity(item_name, new_quantity):
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = new_quantity
            print(f"Updated '{item_name}' quantity to {new_quantity}.")
            return

    print(f"'{item_name}' is not in the cart.")

```

```python
# === Function to PRINT current cart state 
def print_cart():
    if len(cart) == 0:
        print("Cart is empty.")
    else:
        print("Current Cart:", cart)
        

# Step 1: Add Paneer Tikka x2
add_item("Paneer Tikka", 2)
print_cart()
print()

# Step 2: Add Gulab Jamun x1
add_item("Gulab Jamun", 1)
print_cart()
print()

# Step 3: Add Paneer Tikka x1 (should update quantity to 3)
add_item("Paneer Tikka", 1)
print_cart()
print()

# Step 4: Try to add Mystery Burger (does not exist)
add_item("Mystery Burger", 1)
print_cart()
print()

# Step 5: Try to add Chicken Wings (unavailable)
add_item("Chicken Wings", 1)
print_cart()
print()

# Step 6: Remove Gulab Jamun
remove_item("Gulab Jamun")
print_cart()
print()        
```

==
Task 3 — Inventory Tracker with Deep Copy
==

```python
import copy

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}


cart = [
    {"item": "Paneer Tikka", "quantity": 3, "price": 180.0}
]

# 1: Deep copy inventory ---
inventory_backup = copy.deepcopy(inventory)

inventory["Paneer Tikka"]["stock"] = 999

print("\n-- Proving Deep Copy Works --")
print("inventory Paneer Tikka stock    :", inventory["Paneer Tikka"]["stock"])
print("inventory_backup Paneer Tikka stock:", inventory_backup["Paneer Tikka"]["stock"])
print("(backup is unaffected - deep copy works!)")

# Restore inventory to original state
inventory["Paneer Tikka"]["stock"] = 10
print("\nInventory restored to original.")

#-- 2: Simulate order fulfilment ----
print("\n-- Deducting Cart Items from Inventory --")

for entry in cart:
    item_name = entry["item"]
    qty_needed = entry["quantity"]
    available_stock = inventory[item_name]["stock"]

    if available_stock >= qty_needed:
        inventory[item_name]["stock"] = available_stock - qty_needed
        print(f"Deducted {qty_needed} of '{item_name}'. Remaining stock: {inventory[item_name]['stock']}")
    else:
    
        inventory[item_name]["stock"] = 0
        print(f"Warning: Only {available_stock} unit(s) of '{item_name}' available. Deducted {available_stock}, stock now 0.")

# -- 3: Reorder Alerts---
print("\n-- Reorder Alerts --")

for item, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print(f"Reorder Alert: {item} - Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")

# --4: Print both to confirm they differ--
print("\n-- Final Comparison: inventory vs inventory_backup --")
print(f"{'Item':<18} {'Current Stock':>15} {'Backup Stock':>14}")
print("-" * 50)
for item in inventory:
    current = inventory[item]["stock"]
    backup  = inventory_backup[item]["stock"]
    print(f"{item:<18} {current:>15} {backup:>14}")

print("\nDeep copy protected the original backup successfully!")
```

#1: Total revenue per day

```python
# 1: Total revenue per day---

print("\n-- Part 1: Revenue Per Day --")
for date, orders in sales_log.items():
    daily_revenue = 0
    for order in orders:
        daily_revenue = daily_revenue + order["total"]
    print(f"{date}  :  Rs{daily_revenue}")
```


#2: Best-selling day

```python
#2: Best-selling day ---
print("\n-- Part 2: Best-Selling Day --")

best_day = ""
best_revenue = 0

for date, orders in sales_log.items():
    daily_revenue = 0
    for order in orders:
        daily_revenue = daily_revenue + order["total"]
    if daily_revenue > best_revenue:
        best_revenue = daily_revenue
        best_day = date

print(f"Best-selling day: {best_day} with Rs{best_revenue} revenue")
```

#3: Most ordered item 

```python
#--3: Most ordered item ---
print("\n-- Part 3: Most Ordered Item --")

item_count = {}

for date, orders in sales_log.items():
    for order in orders:
        for item in order["items"]:
            if item not in item_count:
                item_count[item] = 0
            item_count[item] = item_count[item] + 1

most_ordered = max(item_count, key=lambda x: item_count[x])
print(f"Most ordered item: {most_ordered} (appeared in {item_count[most_ordered]} orders)")
```

#4: Add new day and reprint

```python
# -- 4: Add new day and reprint--
print("\n-- Part 4: Adding New Day 2025-01-05 --")

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("Updated Revenue Per Day:")
for date, orders in sales_log.items():
    daily_revenue = 0
    for order in orders:
        daily_revenue = daily_revenue + order["total"]
    print(f"{date}  :  Rs{daily_revenue}")

# Reprint best-selling day after adding new data
best_day = ""
best_revenue = 0

for date, orders in sales_log.items():
    daily_revenue = 0
    for order in orders:
        daily_revenue = daily_revenue + order["total"]
    if daily_revenue > best_revenue:
        best_revenue = daily_revenue
        best_day = date

print(f"\nUpdated Best-Selling Day: {best_day} with Rs{best_revenue} revenue")
```

#5: Numbered list of all orders using enumerate

```python
# 5: Numbered list of all orders using enumerate --
print("\n-- Part 5: All Orders (using enumerate) --")

all_orders = []
for date, orders in sales_log.items():
    for order in orders:
        all_orders.append((date, order))

for number, (date, order) in enumerate(all_orders, start=1):
    items_str = ", ".join(order["items"])
    print(f"{number}. [{date}] Order #{order['order_id']} - Rs{order['total']:.2f} - Items: {items_str}")
```
